```r
#Cell 1
#Load package/library
library("qiime2R")

#Import data
#Paths set to "data/XXX" as there is a subdirectory with the data in the same directory as this notebook
pseq <- qiime2R::qza_to_phyloseq(
    features = "data/table-dada2.qza",
    tree = "data/rooted-tree.qza",
    taxonomy = "data/taxonomy.sklearn.qza",
    metadata = "data/media_metadata.txt"
)
```

```r
#Cell 2
#Load microbiome library
library("microbiome")
#Summary of phyloseq object
microbiome::summarize_phyloseq(pseq)
```

```r
#Cell 3
#Save phyloseq as file
save(pseq, file = "phyloseq.RData")
```
